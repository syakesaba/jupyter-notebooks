```python
from geopy.distance import geodesic
import json
import math

json_ = '[{"type":"polyline","latLngs":[{"lat":34.688321,"lng":135.519201},{"lat":34.704693,"lng":135.527484}],"color":"#a24ac3"},{"type":"polyline","latLngs":[{"lat":34.688321,"lng":135.519201},{"lat":34.702246,"lng":135.506086}],"color":"#a24ac3"},{"type":"polyline","latLngs":[{"lat":34.704693,"lng":135.527484},{"lat":34.702246,"lng":135.506086}],"color":"#a24ac3"},{"type":"circle","latLng":{"lat":34.6984145094219,"lng":135.51767823792733},"radius":1139.748138135289,"color":"#a24ac3"}]'
# json_ = '[{"type":"polyline","latLngs":[{"lat":34.688321,"lng":135.519201},{"lat":34.704693,"lng":135.527484}],"color":"#a24ac3"},{"type":"polyline","latLngs":[{"lat":34.688321,"lng":135.519201},{"lat":34.702246,"lng":135.506086}],"color":"#a24ac3"},{"type":"polyline","latLngs":[{"lat":34.704693,"lng":135.527484},{"lat":34.702246,"lng":135.506086}],"color":"#a24ac3"},{"type":"circle","latLng":{"lat":34.6984145094219,"lng":135.51767823792733},"radius":1139.748138135289,"color":"#a24ac3"}]'

draws = json.loads(json_)
polylines = [draw["latLngs"] for draw in draws if draw["type"] == "polyline"]
assert len(polylines) == 3
distances = []
for polyline in polylines:
  assert len(polyline) == 2
  point1 = (polyline[0]["lat"], polyline[0]["lng"])
  point2 = (polyline[1]["lat"], polyline[1]["lng"])
  distance = geodesic(point1, point2).kilometers
  distances.append(distance)
if len(set(distances)) == 1:
  print(f"一辺が{distances[0]}キロメートルの正三角形です！")
else:
  print("***正三角形ではありません***")
  print(f"辺1: {distances[0]}キロメートル")
  print(f"辺2: {distances[1]}キロメートル")
  print(f"辺3: {distances[2]}キロメートル")
  p = sum(distances) / 2
  S = math.sqrt(p * (p-distances[0]) * (p-distances[1]) * (p-distances[2]))
  print(f"周長: {sum(distances)}キロメートル")
  print(f"面積: {S}平方キロメートル")
  t = sum(distances) / len(distances)
  tp = sum(distances) / 2
  St = math.sqrt(tp * (tp-t) * (tp-t) * (tp-t))
  print(f"正三角形の場合の面積: {St}平方キロメートル")
  print(f"想定スコア: {(S/St)*100}%")
```
